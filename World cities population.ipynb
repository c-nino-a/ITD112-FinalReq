```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly
```

```python
dataset = pd.read_csv("./Final Requirements Datasets/World cities population.csv")
```

```python
dataset.head()
```

```python
nan_value = float("NaN")

dataset.replace("", nan_value, inplace=True)

dataset.dropna(subset = ["population"], inplace=True)

df=dataset[dataset["iso2"]=="CN"]

df
```

```python
fig=px.scatter_geo(df,scope="asia",lat="lat",lon="lng",color="admin_name",
                   hover_name="city",size="population", projection="natural earth",labels={"admin_name":"Administration"},
                  center={"lat":34,"lon":101},title="Population bubble map of Chinese Population by administration")
```

```python
fig.show()
```
