```python
graph2 = dataset[['year','month','police_regional_office',"shooting_incidents_total","robbery_total","carnapping_total","others_total","victims_total"]]
graph2["shooting_incidents_total"] = pd.to_numeric(graph2["shooting_incidents_total"],errors='coerce').fillna(0).astype(int)
graph2["robbery_total"] = pd.to_numeric(graph2["robbery_total"],errors='coerce').fillna(0).astype(int)
graph2["carnapping_total"] = pd.to_numeric(graph2["carnapping_total"],errors='coerce').fillna(0).astype(int)
graph2["others_total"] = pd.to_numeric(graph2["others_total"],errors='coerce').fillna(0).astype(int)
graph2["victims_total"] = pd.to_numeric(graph2["victims_total"],errors='coerce').fillna(0).astype(int)

graph2
```

```python
graph2test = pd.DataFrame()


graph2test["total"] = graph2.groupby(['police_regional_office'])['victims_total'].sum()

graph2test["shooting"] = graph2.groupby(['police_regional_office'])['shooting_incidents_total'].sum()

graph2test["carnapping"] = graph2.groupby(['police_regional_office'])['carnapping_total'].sum()

graph2test["robbery"] = graph2.groupby(['police_regional_office'])['robbery_total'].sum()

graph2test["others"] = graph2.groupby(['police_regional_office'])['others_total'].sum()
```

```python
graph2test=graph2test.reindex(["1","2","3","4A","4B","5","6","7","8","9","10","11","12","13","ARMM","COR","NCRPO"])
graph2test
```

```python
column_names = ["shooting","carnapping","robbery","others","total"]

graph2test = graph2test.reindex(columns=column_names)
```

```python
graph2test=graph2test.sort_values("total",axis=0,ascending=False)
graph2test = graph2test.drop("total",axis=1)
graph2test
```
