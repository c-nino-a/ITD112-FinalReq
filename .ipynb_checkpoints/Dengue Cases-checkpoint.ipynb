```python
months=dataset.Month.unique()
years=dataset.Year.unique()
regions=dataset.Region.unique()
splitsets={}

dataset['Year'] = dataset.Year.astype(str)

# for year in years:
currentyear=dataset.loc[dataset['Year'] == "2016"]
for region in regions:
    currentregion=currentyear.loc[currentyear['Region']==region]
    name="2016"+"_"+str(region)
    splitsets[name]=currentregion

```

```python
currentyear
```

```python
import more_itertools

splitsets
```

```python
ax=splitsets["2016_Region.I"].plot.line(x="Month",y="Dengue_Cases",label="Region.I")

for frame in splitsets:
    splitsets[frame].plot.line(figsize=(10,10),ax=ax,x="Month",y="Dengue_Cases",label=frame[5:],xlabel="Months of 2016",ylabel="Number of Dengue cases",title="Number of dengue cases by region in the year 2016")
```

```python
regionx2016=splitsets["2016_Region.X"].sort_values("Dengue_Cases",axis=0,ascending=False)
regionx2016=regionx2016.set_index(regionx2016["Month"])
regionx2016
```

```python
plot = regionx2016.plot.pie(autopct='%1.1f%%',y="Dengue_Cases",ylabel="",figsize=(10, 10),title="Proportions of dengue cases by month in Region X year 2016",legend=False)
```
