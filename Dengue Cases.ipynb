```python
months=dataset.Month.unique()
years=dataset.Year.unique()
regions=dataset.Region.unique()
splitsets={}

# dataset['Year'] = dataset.Year.astype(str)

# for year in years:
currentyear=dataset.loc[dataset['Year'] == "2016"]
for region in regions:
    currentregion=currentyear.loc[currentyear['Region']==region]
    name=str(year)+"_"+str(region)
    splitsets[name]=currentregion

```
